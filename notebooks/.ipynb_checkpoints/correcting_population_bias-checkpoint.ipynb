# Summary Introduction

In this analysis, we aim to correct biases in a sample dataset (`demographic_attributes.csv`) to better reflect the ground truth distribution provided in a ground truth dataset (`demo_ground_truth.csv`). This correction is essential for ensuring that the sample data accurately represents various demographic groups, including age, education, and ethnicity. By calculating and applying weights, we adjust for any discrepancies between the sample and ground truth distributions. This process involves several key steps: loading and inspecting the data, computing sample and ground truth distributions, calculating weights, and applying these weights to correct the sample data. The final output is a CSV file with adjusted weights that better align with the ground truth distribution.

### GT Age

```python
# Custom sorting function
def sort_key(age_group):
    """
    Custom sorting function for age groups represented as strings.

    This function takes an age group string formatted as 'X_Y', where X and Y are integers representing 
    the lower and upper bounds of the age range, respectively. It splits the string into its components 
    and converts them into integers, returning a tuple that can be used for sorting.

    Parameters:
    age_group (str): A string representing an age group in the format 'X_Y'.

    Returns:
    tuple: A tuple of integers (X, Y) that can be used as a sorting key.
    """
    # Split the string and convert to integers
    return tuple(map(int, age_group.split('_')))

age_list = ['0_4', '5_9', '10_14', '15_17', '18_24', '25_34', 
            '35_44', '45_54', '55_64', '65_74', '75_84', '85_120']
age_ground_truth_df = (
    demographic_ground_truth_df[demographic_ground_truth_df['demographic category'].isin(age_list)]
    .rename(columns={'demographic category': 'age'})
    .set_index('age')
    .sort_index(key=lambda x: x.map(sort_key))
)
```

# Correcting Sample Data Bias

The objective of this project is to correct the bias in the sample dataset (`demographic_attributes.csv`) to better reflect the ground truth distribution (`demo_ground_truth.csv`). This can be achieved by calculating weights that adjust for the over- or under-representation of groups in the sample data. This involves calculating the distribution of demographic categories in both the sample and ground truth data. By comparing these distributions, we can compute weights to adjust for any over- or under-representation in the sample data. 

Steps

1. Calculate the sample distributions for age, education, and ethnicity from the sample data
    - This is done using the value_counts() method in pandas, which provides the count of unique values.
2. Calculate the ground truth distributions
    - Using the ground truth data provided in demo_ground_truth.csv, calculate the true distribution of the same categories.
3. Construct Weights
    - Calculate the weights for each category by dividing the ground truth distribution by the sample distribution. This helps to correct the bias by adjusting for the over- or under-representation of groups in the sample data.
4. Impute Missing Age Groups and Calculate Age Weights
    - Since some age groups are missing in the sample data, impute the missing age group counts based on proportions of the ground truth data and then calculate the weights.
5. Apply Weights to the Sample Data
    - Map the calculated weights back to the sample data and compute the combined weight for each individual.
6. Save the Adjusted Weights to CSV

# Calculate Weights For Correcting Sample Data

The following code is designed to compute and normalize demographic weights based on sample and ground truth distributions for various categories such as education, ethnicity, and age.

The `compute_weights` function calculates weights for a given demographic category by dividing the ground truth distribution by the sample distribution.

The `impute_missing_age_groups` function identifies age groups present in the ground truth distribution but missing in the sample data. It imputes these missing counts based on the proportions from the ground truth data.

The `calculate_age_weights` function utilizes the imputed age groups to compute the adjusted sample age distribution. It then calculates weights for each age group by dividing the ground truth distribution by the adjusted sample distribution.


#### Helper Method For Computing Weights For Education and Ethnicity

```python
def compute_weights(demographic_category, sample_distribution, ground_truth_distribution):
    """
    Compute weights for a given demographic category based on the sample distribution and ground truth distribution.

    Parameters:
    demographic_category (str): The demographic category for which weights are being computed (not used in the calculation).
    sample_distribution (pd.Series): A Series representing the observed distribution of the demographic category in the sample.
    ground_truth_distribution (pd.Series): A Series representing the expected distribution of the demographic category based on ground truth data.

    Returns:
    pd.Series: A Series containing the calculated weights for each demographic category, where weights are computed as the ratio of ground truth distribution to the sample distribution.
    """
    weights = ground_truth_distribution / sample_distribution.reindex(ground_truth_distribution.index, fill_value=1e-10)
    return weights
```

#### Compute Weights For Education and Ethnicity

```python
education_weights = compute_weights('education', sample_education_distribution, ground_truth_education_distribution)
ethnicity_weights = compute_weights('ethnicity', sample_ethnicity_distribution, ground_truth_ethnicity_distribution)

# Merge weights back to the original dataframe
demographic_attributes_df['education_weight'] = demographic_attributes_df['education'].map(education_weights)
demographic_attributes_df['ethnicity_weight'] = demographic_attributes_df['ethnicity'].map(ethnicity_weights)
```

#### Helper Method For Imputing Missing Age Groups & Generating Weights

```python
def impute_missing_age_groups(demographic_attributes_df, ground_truth_age_distribution):
    """
    Impute missing age group counts in the demographic data based on ground truth proportions.

    Parameters:
    demographic_attributes_df (pd.DataFrame): DataFrame containing demographic attributes including age.
    ground_truth_age_distribution (pd.Series): Series representing the ground truth age distribution.

    Returns:
    pd.Series: A Series containing the imputed counts for missing age groups.
    """
    sample_age_distribution = demographic_attributes_df['age'].value_counts()
    total_sample_individuals_data = len(demographic_attributes_df)
    
    # Identify missing age groups
    missing_age_groups = set(ground_truth_age_distribution.index) - set(demographic_attributes_df['age'].unique())
    missing_age_groups = list(missing_age_groups)
    
    # Impute counts for missing age groups based on ground truth proportions
    imputed_counts = ground_truth_age_distribution[missing_age_groups] * total_sample_individuals_data
    
    # Combine the original and imputed counts using pd.concat
    sample_age_distribution_full = pd.concat([sample_age_distribution, imputed_counts]).sort_index()
    
    return sample_age_distribution_full


def calculate_age_weights(demographic_attributes_df, ground_truth_age_distribution):
    """
    Calculate weights for age groups based on the demographic data and ground truth distribution.

    Parameters:
    demographic_attributes_df (pd.DataFrame): DataFrame containing demographic attributes including age.
    ground_truth_age_distribution (pd.Series): Series representing the ground truth age distribution.

    Returns:
    pd.Series: A Series containing the calculated weights for each age group.
    """
    sample_age_distribution_full = impute_missing_age_groups(demographic_attributes_df, ground_truth_age_distribution)
    
    # Calculate the adjusted sample distribution
    adjusted_sample_age_distribution = sample_age_distribution_full / sample_age_distribution_full.sum()
    
    # Calculate weights for age groups
    age_weights = ground_truth_age_distribution / adjusted_sample_age_distribution.reindex(ground_truth_age_distribution.index, fill_value=0)
    
    return age_weights
```

```python
age_weights = calculate_age_weights(demographic_attributes_df, ground_truth_age_distribution)
demographic_attributes_df['age_weight'] = demographic_attributes_df['age'].map(age_weights)
```

#### Notes:
- For resolving the missing age groups in the sample data, I initially looked at three different methods, namely the above `impute_missing_age_groups` method, and also the following `smooth_regularization_adjustment` and `synthetic_data_generation` methods.
- After comparing the `age_weights` generated for all three, as well as comparing runtimes for each, I adopted the use of `impute_missing_age_groups` because of its faster execution time and comparable results to those of `sythetic_data_generation`.
- Justifications for this can be found in the code cells below.

```python
def smooth_regularization_adjustment(demographic_attributes_df, ground_truth_age_distribution, epsilon=0.05):
    """
    Adjust the sample age distribution using smooth regularization to avoid zero counts. 

    This method adds a small constant (epsilon) to both the sample age counts and the total sample size 
    before calculating proportions. This prevents division by zero errors and smooths the distribution.

    Parameters:
    demographic_attributes_df (pd.DataFrame): DataFrame containing demographic attributes including age.
    ground_truth_age_distribution (pd.Series): Series representing the ground truth age distribution.
    epsilon (float): Small value added to avoid division by zero (default is 0.05).

    Returns:
    pd.Series: A Series containing the calculated weights for each age group based on the adjusted distribution.
    """
    sample_age_distribution = demographic_attributes_df['age'].value_counts()
    number_of_individuals_in_sample_data = len(demographic_attributes_df)
    
    # Adjust the sample distribution to avoid zeros
    adjusted_sample_age_distribution = (sample_age_distribution + epsilon) / (number_of_individuals_in_sample_data + epsilon * len(sample_age_distribution))
    
    # Calculate weights for age groups
    age_weights = ground_truth_age_distribution / adjusted_sample_age_distribution.reindex(ground_truth_age_distribution.index, fill_value=epsilon)
    
    return age_weights


def synthetic_data_generation(demographic_attributes_df, ground_truth_age_distribution):
    """
    Generate synthetic data entries for missing age groups based on the ground truth distribution.

    This method identifies age groups that are absent in the sample data and creates synthetic data entries
    for those groups to match the proportions found in the ground truth distribution. 

    Parameters:
    demographic_attributes_df (pd.DataFrame): DataFrame containing demographic attributes including age.
    ground_truth_age_distribution (pd.Series): Series representing the ground truth age distribution.

    Returns:
    pd.Series: A Series containing the calculated weights for each age group based on the synthetic data generated.
    """
    sample_age_distribution = demographic_attributes_df['age'].value_counts()
    number_of_individuals_in_sample_data = len(demographic_attributes_df)
    
    # Identify missing age groups
    missing_age_groups = set(ground_truth_age_distribution.index) - set(demographic_attributes_df['age'].unique())
    
    # Create synthetic entries for missing age groups
    synthetic_entries = []
    for age_group in missing_age_groups:
        synthetic_entries.extend([{'person id': np.nan, 
                                   'age': age_group, 
                                   'education': 'Unknown', 
                                   'ethnicity': 'Unknown'}] * int(ground_truth_age_distribution[age_group] * number_of_individuals_in_sample_data))
    
    synthetic_df = pd.DataFrame(synthetic_entries)
    
    # Append synthetic entries to the sample data
    demographic_attributes_df_full = pd.concat([demographic_attributes_df, synthetic_df], ignore_index=True)
    
    # Recalculate the sample distribution
    sample_age_distribution_full = demographic_attributes_df_full['age'].value_counts() / len(demographic_attributes_df_full)
    
    # Calculate weights for age groups
    age_weights = ground_truth_age_distribution / sample_age_distribution_full.reindex(ground_truth_age_distribution.index, fill_value=0)
    
    return age_weights

def imputation_with_ground_truth_proportions(demographic_attributes_df, ground_truth_age_distribution):
    """
    Impute missing age group counts in the demographic data based on ground truth proportions.

    Parameters:
    demographic_attributes_df (pd.DataFrame): DataFrame containing demographic attributes including age.
    ground_truth_age_distribution (pd.Series): Series representing the ground truth age distribution.

    Returns:
    pd.Series: A Series containing the calculated weights for each age group based on the imputed counts.
    """
    sample_age_distribution = demographic_attributes_df['age'].value_counts()
    number_of_individuals_in_sample_data = len(demographic_attributes_df)
    
    # Identify missing age groups
    missing_age_groups = set(ground_truth_age_distribution.index) - set(demographic_attributes_df['age'].unique())
    missing_age_groups = list(missing_age_groups)
    
    # Impute counts for missing age groups
    imputed_counts = ground_truth_age_distribution[missing_age_groups] * number_of_individuals_in_sample_data
    
    # Combine the original and imputed counts using pd.concat
    sample_age_distribution_full = pd.concat([sample_age_distribution, imputed_counts]).sort_index()
    
    # Calculate the adjusted sample distribution
    adjusted_sample_age_distribution = sample_age_distribution_full / sample_age_distribution_full.sum()
    
    # Calculate weights for age groups
    age_weights = ground_truth_age_distribution / adjusted_sample_age_distribution.reindex(ground_truth_age_distribution.index, fill_value=0)
    
    return age_weights
```

```python
import numpy as np

# Generate age weights from each function
weights_smooth = smooth_regularization_adjustment(demographic_attributes_df, ground_truth_age_distribution)
weights_imputation = imputation_with_ground_truth_proportions(demographic_attributes_df, ground_truth_age_distribution)
weights_synthetic = synthetic_data_generation(demographic_attributes_df, ground_truth_age_distribution)

# Create a single figure for overlapping scatter plots
plt.figure(figsize=(10, 6))

# Define the x locations for the age groups
x = np.arange(len(weights_smooth.index))  # X locations for the groups

# Plot each set of weights as dots with different shapes and transparency
plt.scatter(x, weights_smooth.values, color='blue', alpha=0.6, label='Smooth Regularization Adjustment; epsilon=0.05', s=100, marker='o')  # Circle
plt.scatter(x, weights_imputation.values, color='red', alpha=0.6, label='Imputation with Ground Truth Proportions', s=100, marker='s')  # Square
plt.scatter(x, weights_synthetic.values, color='green', alpha=0.6, label='Synthetic Data Generation', s=100, marker='^')  # Triangle

# Set titles and labels
plt.title('Age Weights from Different Methods')
plt.ylabel('Weights')
plt.xlabel('Age Groups')
plt.xticks(x, weights_smooth.index, rotation=45)  # Set x-ticks to age groups
plt.legend()  # Show legend

# Adjust layout
plt.tight_layout()
plt.show()
```

## Calculate Combined Weight For Each Person

Normalization is performed to ensure that the calculated weights are scaled appropriately. This involves adjusting the weights for each category (age, education, ethnicity) based on the total number of individuals in the sample and ground truth data.

A scaling factor is applied to the combined normalized weights to bring them into a more interpretable range.

```python
# Normalize age weights
demographic_attributes_df['age_weight_normalized'] = (
    demographic_attributes_df['age_weight'] * total_sample_individuals / total_ground_truth_age
)

# Normalize education weights
demographic_attributes_df['education_weight_normalized'] = (
    demographic_attributes_df['education_weight'] * total_sample_individuals / total_ground_truth_education
)

# Normalize ethnicity weights
demographic_attributes_df['ethnicity_weight_normalized'] = (
    demographic_attributes_df['ethnicity_weight'] * total_sample_individuals / total_ground_truth_ethnicity
)

# Calculate combined normalized weight
demographic_attributes_df['weight_normalized'] = (
    demographic_attributes_df['age_weight_normalized'] *
    demographic_attributes_df['education_weight_normalized'] *
    demographic_attributes_df['ethnicity_weight_normalized']
)

# Apply scaling factor to weights
scaling_factor = 1e8
demographic_attributes_df['weight'] = demographic_attributes_df['weight_normalized'] * scaling_factor
```
